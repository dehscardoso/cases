Agora, será aplicada a técnica de **feature engineering**, que é um processo de usar o conhecimento do domínio para criar novas características (features) a partir dos dados brutos, com o objetivo de melhorar o desempenho dos modelos de aprendizado de máquina. Esse processo é crucial para garantir que os modelos possam extrair o máximo de informação útil dos dados.

```python
# Calcular o custo total do cliente dividido pelo tempo de permanência
churn_data['TotalChargesPerTenure'] = churn_data['TotalCharges'] / (churn_data['tenure'] + 1e-5)

# Indicar se o cliente tem algum serviço de internet (1) ou não (0)
churn_data['HasInternetService'] = churn_data['InternetService'].apply(lambda x: 0 if x == 'No' else 1)

# Indicar se o cliente tem múltiplas linhas de telefone (1) ou não (0)
churn_data['HasMultipleLines'] = churn_data['MultipleLines'].apply(lambda x: 0 if x == 'No' else 1)

# Contar o número total de serviços adicionais que o cliente possui
churn_data['ServicesCount'] = (
    churn_data[['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']]
    .apply(lambda x: x.apply(lambda y: 0 if y in ['No', 'No internet service'] else 1).sum(), axis=1)
)

# Criar grupos de tempo de permanência (Tenure)
bins = [0, 12, 24, 48, 60, 72]
labels = ['0-12', '12-24', '24-48', '48-60', '60-72']
churn_data['TenureGroup'] = pd.cut(churn_data['tenure'], bins=bins, labels=labels, include_lowest=True)

# Criar uma coluna para indicar se o cliente é idoso e possui dependentes
churn_data['SeniorDependent'] = (churn_data['SeniorCitizen'] == 1) & (churn_data['Dependents'] == 'Yes')
churn_data['SeniorDependent'] = pd.to_numeric(churn_data['SeniorDependent'].replace({True: 1, False: 0}), errors='coerce')

# Contar o número de serviços adicionais assinados pelo cliente
additional_services = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']
churn_data['ServicesCount'] = churn_data[additional_services].apply(lambda x: (x == 'Yes').sum(), axis=1)

# Converter o tipo de contrato para uma característica numérica
contract_mapping = {'Month-to-month': 1, 'One year': 12, 'Two year': 24}
churn_data['ContractType'] = churn_data['Contract'].map(contract_mapping)

# Convertendo colunas
churn_data['Partner'] = pd.to_numeric(churn_data['Partner'].replace({'Yes': 1, 'No': 0}), errors='coerce')
churn_data['Dependents'] = pd.to_numeric(churn_data['Dependents'].replace({'Yes': 1, 'No': 0}), errors='coerce')
churn_data['PhoneService'] = pd.to_numeric(churn_data['PhoneService'].replace({'Yes': 1, 'No': 0}), errors='coerce')
churn_data['MultipleLines'] = pd.to_numeric(churn_data['MultipleLines'].replace({'Yes': 1, 'No': 0}), errors='coerce')
churn_data['PaperlessBilling'] = pd.to_numeric(churn_data['PaperlessBilling'].replace({'Yes': 1, 'No': 0}), errors='coerce')
churn_data['gender'] = pd.to_numeric(churn_data['gender'].replace({'Male': 1, 'Female': 0}), errors='coerce')
```

```python
# Coluna a ser excluída da codificação
col_to_exclude = 'Churn'

# Colunas a serem codificadas
cols_to_encode = churn_data.columns.difference([col_to_exclude])

# Codificar as colunas selecionadas
encoded_data = pd.get_dummies(churn_data[cols_to_encode], drop_first=True)

# Combinar com a coluna não codificada
final_data = pd.concat([encoded_data, churn_data[[col_to_exclude]]], axis=1)

# Convertendo Churn
final_data['Churn'] = pd.to_numeric(final_data['Churn'].replace({'Yes': 1, 'No': 0}), errors='coerce')
```

```python
# Definir variáveis independentes e dependentes
features = ['ContractType', 'Dependents', 'HasInternetService', 'HasMultipleLines',
            'MonthlyCharges', 'MultipleLines', 'PaperlessBilling', 'Partner',
            'PhoneService', 'SeniorCitizen', 
            'SeniorDependent', 'ServicesCount', 'TotalCharges',
            'TotalChargesPerTenure', 'gender', 'tenure', 'Contract_One year',
            'Contract_Two year', 'DeviceProtection_No internet service',
            'DeviceProtection_Yes', 'InternetService_Fiber optic', 'InternetService_No', 
            'OnlineBackup_No internet service', 'OnlineBackup_Yes',
            'OnlineSecurity_No internet service', 'OnlineSecurity_Yes',
            'PaymentMethod_Credit card (automatic)', 'PaymentMethod_Electronic check', 
            'PaymentMethod_Mailed check', 'StreamingMovies_No internet service', 'StreamingMovies_Yes',
            'StreamingTV_No internet service', 'StreamingTV_Yes',
            'TechSupport_No internet service', 'TechSupport_Yes',
            'TenureGroup_12-24', 'TenureGroup_24-48', 'TenureGroup_48-60', 'TenureGroup_60-72']

target = 'Churn'

X = final_data[features]
y = final_data[target]
```
